```
labels = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/earthquake/train_labels.csv')
labels = labels['damage_grade']
```

###Handling categorical features

```
def cat (data):
  categories = []
  #collecting all the categorical features
  for x in data.columns:
    if data[str(x)].dtype == 'object':
      categories.append(x)
  #using only the top 5 unique values and replacing other values with 'others' 
  #in each categorical feature to resuce dimensionality of the final data
  for x in categories:
    if len(data[x].unique()) > 5:
      y = data[x].replace(data[x].unique()[5:], 'others')
      print(len(y))
      data[x] = y
  return data
```

```
data_ = cat(data)
```

```
data_
```

```
data_ = pd.get_dummies(data_, drop_first=True)
```

```
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```
x_train, x_test , y_train, y_test = train_test_split(data_, labels, test_size=0.2)
```

```
from sklearn.ensemble import RandomForestClassifier
```

```
model_for = RandomForestClassifier()
```

```
model_for.fit(x_train, y_train)
```

```
preds = model_for.predict(x_test)
```

```
print(accuracy_score(y_test, preds))
```
